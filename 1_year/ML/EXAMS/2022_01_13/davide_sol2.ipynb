 ### 3.tune the hyper–parameters of Model1 with Cross Validation on the training set, optimize for recall_macro

```python
y = df[target].copy()
X = df.drop(target, axis=1).copy()
```

```python
#Train and test
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y
                                                    , train_size = 0.66
                                                    , random_state = random_state) # default Train 0.75- Test 0.25
print("There are {} samples in the training dataset".format(X_train.shape[0]))
print("There are {} samples in the testing dataset".format(X_test.shape[0]))
print("Each sample has {} features".format(X_train.shape[1]))
```

Decision Tree

```python
#Define just one type of classifier: decision tree in this case
#change the params and estimator with the choices listed in the previous cell
#gridSearchCv uses CrossValidation
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV

model_param_dt = {'criterion':['gini', 'entropy'], 'max_depth':list(range(1,10)), 'min_samples_split': range(2,10)}
estimator_dt = DecisionTreeClassifier()
score = 'recall_macro'


model1 = GridSearchCV(estimator_dt, model_param_dt, cv=5,
                   scoring=score,
                   return_train_score=False,
                   n_jobs=2,  # this allows using multi-cores
                   )

model1.fit(X_train, y_train)
print("best parameter estimator :", model1.best_params_)
print("best score for recall_macro: {:.2f}%".format(model1.best_score_*100))
print()

#Then predict the labels
y_true, y_pred_dt = y_test, model1.predict(X_test)

#Evaluate accuracy of the model on the test set
from sklearn.metrics import accuracy_score
print("the accuracy score is: ",round(accuracy_score(y_test,y_pred_dt)*100, 2),'%' )

```

## 4.produce a classification report for Model1 on the test set

```python
from sklearn.metrics import classification_report

print(classification_report(y_true, y_pred_dt))
```
